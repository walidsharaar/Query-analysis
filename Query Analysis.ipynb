```python
#import pandas
import pandas as pd
from ipywidgets import interact
```

```python
#import data
data= pd.read_csv("D:/Analytics/employee.csv")
data.head()
```

```python
#group the employee details by Department

data.groupby(data['Department']).agg('mean')
```

```python
# group the employee based on department and mean of age

data[['Age','Department']].groupby(data['Department']).agg('mean')
```

```python
# group the employee based on department and min of age

data[['Age','Department']].groupby(data['Department']).agg('min')
```

```python
# group the employee based on department and std of age

data[['Age','Department']].groupby(data['Department']).agg('std')
```

```python
# group the employee based on department and variance of age

data[['Age','Department']].groupby(data['Department']).agg('var')
```

```python
# group the employee based on department and count of age

data[['Age','Department']].groupby(data['Department']).agg('count')
```

```python
# group the employee based on department and count of age

data[['Age','Department']].groupby(data['Department']).agg('nunique')
```

```python
# idxmax and idxmin function
# number of employee which highest salary

data.loc[data.groupby(data['Department'])['MonthlyIncome'].idxmax()][['EmployeeNumber','Department','MonthlyIncome']]
```

```python
# employees with less amount of salary

data.loc[data.groupby(data['Department'])['MonthlyIncome'].idxmin()][['EmployeeNumber','Department','MonthlyIncome']]
```

```python
# color scales with groupby

data[['Age','Department']].groupby(data['Department']).agg('mean').style.background_gradient(cmap='copper')
```

```python
data[['Age','Department']].groupby(data['Department']).agg('mean').style.background_gradient(cmap='Reds')
```

```python
data[['MonthlyIncome','Department']].groupby(data['Department']).agg('mean')
```

```python
data[['MonthlyIncome','Department']].groupby(data['Department']).agg('mean').plot(kind='line')
```

# pivottables

The general syntax is pivot_table(data,values= None, index-None, column=None, aggfunc='mean',...)

- data is dataframe
- values contains the column to aggregate
- index is the row in the pivot table
- columns contains the columns we want in the pivot table
- aggfunc is the aggregate function


```python
# pivottable for the department and their mean ages

data.pivot_table(values='MonthlyIncome', index='Department', aggfunc='mean')
```

```python
#  colorscales using pivottable
data.pivot_table(values='MonthlyIncome', index='Department', aggfunc='mean').style.background_gradient(cmap='copper')
```

```python
data.pivot_table(values='MonthlyIncome', index='Department', aggfunc='mean').plot(kind='line')
```

# cross tabulation

- is a method to quantitavely analyze the relationship between several variables.

```python
data[['Attrition','Department']].head()
```

```python
pd.crosstab(data['Attrition'],data['Department'])
```

```python
# visualizing the cross tab data 

pd.crosstab(data['Attrition'],data['Department']).plot(kind='bar', stacked= True)
```

```python
# Interact the groupby

@interact
def groupby(column = list(data.select_dtypes('object').columns)):
    return data[['Age',column]].groupby([data[column]]).agg(['mean'])
```

```python
@interact
def groupby(column = list(data.select_dtypes('object').columns),
           index= list (data.select_dtypes('number').columns)):
    return data[[index,column]].groupby([data[column]]).agg(['mean'])
```
